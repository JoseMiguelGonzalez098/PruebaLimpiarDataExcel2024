```python
import pandas as pd
import unidecode
import re
```

```python
file = './Challenge Ask A Manager Salary Survey 2021.xlsx'
df = pd.read_excel(file)
```

```python
df.head()
```

```python
def clean_data(value):
    if isinstance(value, str):
        value = value.upper()  # Convertir a mayúsculas
        value = unidecode.unidecode(value)  # Eliminar acentos
        value = re.sub(r'\s+', ' ', value)  # Eliminar espacios múltiples
        value = re.sub(r'\s([?.!,"](?:\s|$))', r'\1', value)  # Corregir espaciado
    return value
```

```python
df.columns = [clean_data(col) for col in df.columns]
```

```python
data_limpia = df.applymap(clean_data)
```

```python
data_limpia.head()
```

```python
file_result = './Challenge Ask A Manager Salary Survey 2021 Clean.xlsx'
data_limpia.to_excel(file_result, index=False)
```
